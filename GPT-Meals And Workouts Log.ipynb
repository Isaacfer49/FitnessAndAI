```python

import re
from collections import defaultdict

def get_daily_calorie_averages(text: str):
    """
    Parses the text for date lines (e.g. '17-August-2024'), then collects
    the lines in each 'Meals' block for that date. 
    Extracts numbers from 'XXX cals' or 'XXX-YYY cals' lines, 
    computes the average for each day, and returns a dictionary:
    {
      "17-August-2024":  (average cals found),
      "18-August-2024":  ...
      ...
    }
    """
    # Regex for matching a date line like "17-August-2024"
    # (start of line, 1-2 digits, dash, word-chars for month, dash, 4 digits)
    date_pattern = re.compile(r"^(\d{1,2}-[A-Za-z]+-\d{4,5})\s*$")

    # Regex for extracting cals, e.g. "245 cals" or "800-900 cals"
    cals_pattern = re.compile(r"(\d+(?:-\d+)?)\s*cals", re.IGNORECASE)

    lines = text.splitlines()
    current_day = None
    collecting_meals = False

    # day -> list of numeric calorie values found
    day_to_cals = defaultdict(list)

    for raw_line in lines:
        line = raw_line.strip()

        # 1) Check if it's a new date line
        if date_pattern.match(line):
            current_day = line
            collecting_meals = False  # reset
            continue

        # 2) If we see "Meals" (case-insensitive), we start collecting
        if line.lower().startswith("meals"):
            # Turn on collecting only if we have a valid current_day
            if current_day is not None:
                collecting_meals = True
            continue

        # 3) If we hit an empty line (or something that might indicate a break),
        #    turn off collecting
        if not line:
            collecting_meals = False
            continue

        # 4) If we're collecting meals for current_day, parse cals from the line
        if collecting_meals and current_day is not None:
            matches = cals_pattern.findall(line)
            if matches:
                for match in matches:
                    # match could be "245" or "800-900"
                    if "-" in match:
                        # "800-900" -> average
                        left, right = match.split("-")
                        try:
                            val = (int(left) + int(right)) / 2.0
                            day_to_cals[current_day].append(val)
                        except ValueError:
                            pass
                    else:
                        # single number e.g. "245"
                        try:
                            val = float(match)
                            day_to_cals[current_day].append(val)
                        except ValueError:
                            pass

    # 5) Compute an average for each day
    day_averages = {}
    for day, cals_list in day_to_cals.items():
        if cals_list:
            avg_cals = sum(cals_list) / len(cals_list)
            day_averages[day] = round(avg_cals, 1)

    return day_averages

# ------------------------------------------------------------------------
# Example usage (in real usage, you'd import this function and then do):
if __name__ == "__main__":
    text_data = """PUT YOUR ENTIRE TEXT ABOVE HERE"""
    daily_avgs = get_daily_calorie_averages(text_data)
    # Print results
    for d in sorted(daily_avgs.keys()):
        print(f"{d}: {daily_avgs[d]} cals (daily average)")

```

```python
import re
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns

def parse_workout_categories(text: str):
    """
    Scans all lines in the text (ignoring 'Meals' blocks and date lines)
    and increments counters for categories (e.g., 'Boxing', 'Pull', 'Run', etc.)
    based on keywords we detect in each line.

    Returns a dict like:
        {
          "boxing": 12,
          "run": 10,
          "pull": 5,
          "push": 7,
          ...
        }
    """

    # Regex to detect a date line like "20-July-2024" or "20-July-20224"
    date_pattern = re.compile(r"^(\d{1,2}-[A-Za-z]+-\d{4,5})\s*$", re.IGNORECASE)

    # We'll skip lines explicitly starting with "Meals" or blank or date lines.

    # Define categories we want to track. Each category can have multiple synonyms.
    # Feel free to adjust or expand these!
    synonyms_map = {
        "pull":       ["pull"],
        "push":       ["push"],
        "boxing":     ["boxing", "box", "kicboxing", "kicbboxing", "sparring"],
        "swim":       ["swim", "swimming"],
        "run":        ["run", "running", "race"], 
        "legs":       ["leg", "legs"],
        "chest":      ["chest"],
        "back":       ["back"],
        "shoulders":  ["shoulders"],
        "abs":        ["abs"],
        "bike":       ["bike", "biking", "cycling"],
        "hiit":       ["hiit"],
        "racquetball":["racquetball"],
        "interval":   ["interval", "intervals"],
    }

    # Build a quick lookup: word -> category
    word_to_category = {}
    for cat, synonyms in synonyms_map.items():
        for s in synonyms:
            word_to_category[s.lower()] = cat

    category_counts = defaultdict(int)

    lines = text.splitlines()
    for raw_line in lines:
        line = raw_line.strip().lower()

        # Skip if blank, or if it starts with "meals", or if it looks like a date
        if not line:
            continue
        if line.startswith("meals"):
            continue
        if date_pattern.match(line):
            continue

        # Split line into tokens by whitespace or commas
        tokens = re.split(r"[,\s]+", line)
        for t in tokens:
            # Remove punctuation from token
            t_clean = re.sub(r"[^\w]", "", t)
            if t_clean in word_to_category:
                cat = word_to_category[t_clean]
                category_counts[cat] += 1

    return dict(category_counts)


def plot_category_counts(category_dict):
    """
    Takes a dict {category: count}, and plots a horizontal bar chart
    sorted by highest count first, using Seaborn for aesthetics.
    """
    # Convert to a sorted list of (category, count) by descending frequency
    items = sorted(category_dict.items(), key=lambda x: x[1], reverse=True)
    categories = [x[0].capitalize() for x in items]  # e.g. 'boxing' -> 'Boxing'
    counts = [x[1] for x in items]

    # Seaborn styling
    sns.set(style="whitegrid", font_scale=1.15)
    plt.figure(figsize=(8, 6))

    # Horizontal bar chart
    ax = sns.barplot(x=counts, y=categories, palette="Spectral")
    ax.set_title("Workout Categories: Overall Frequency Q4 2024", fontsize=16, fontweight='bold')
    ax.set_xlabel("Frequency", fontsize=13)
    ax.set_ylabel("Workout Category", fontsize=13)

    # Label each bar with its count
    for i, v in enumerate(counts):
        plt.text(v + 0.4, i, str(v), va='center', fontweight='bold', fontsize=9)
    plt.tight_layout()
    plt.savefig("top_workouts.png", dpi=300, bbox_inches='tight')
    plt.show()


# -------------------- EXAMPLE USAGE --------------------
if __name__ == "__main__":
    text_data = """23-August-2024

1,000 meters swim


Meals:



Oatmeal w Banana
Half sandwich
Protein shake
Sandwich 
2 granola bars
Chicken w soup
1 quesadilla w turkey ham


01-September-2024

Half a marathon 

03-September-2024

1050 swim


04-September-2024

Back workout

05-September-2024

Chest
Boxing	

Meals:

3 Boiled eggs w turkey slices
Beef w Rice
Small portion of cake
Protein shake
Granola bar

06-September-2024

5 km 29:47
1,400 swim

Meals:

Oatmeal w Banana
3 boiled egg whites 1 yolk
2 quesadillas w just one slice of cheese 
Protein shake
Rice w Beef
Granola Bar


07-September-2024
Nothing
Ate terrible bad

08-September-2024

12 kms zone 2
Ate good
Music festival so I ate a burger

09-September-2024

Rest day

Ate terribly bad

10-September-2024

Boxing session

Meals

Oatmeal w Banana
One quesadilla with turkey ham
Protein shake
Beef w French fries 
One quesadilla with turkey ham

11-September-2024

Intervals
Swimming sesh


Meals
3 WHITES 1 YOLK boiled eggs
2 quesadillas
Protein shake w Banana
Pasta w Pork meat w and a no sugar coke
Granola Bar
Pork meat w tortilla
1 cupcake

12-September-2024

Boxing

Meals
3 WHITES 1 YOLK boiled eggs
2 quesadillas
Water  + milk protein shake
Pasta w Pork meat w and a no sugar coke
Granola Bar
Tuna w vegetables and rice


13-September-2024

Swimming sesh
Push



14-September-2024

Pull session
Light boxing session

Meals:  Really bad

15-September-2024

10 km trail run zone 2

Meals:

Really bad

16-September-2024

Chest session 
1200 swim (intervals) 


Meals:

Activia yogurt
Half a banana 
Electrolyte
Egg whites panini
Decaf cappuccino 
Protein shake

17-September-2024

Boxing sparring session


Meals:

Protein shake w banana


18-September-2024

Intervals
1,350m swim

Meals

Orange juice
Decaf coffee
3 egg whites
Protein shake
Chicken + chicharron
Chicharron + quesadilla


19-September-2024

Rest
20-September-2024

Boxing

Meals: Bad


21-September-2024

Boxing
Shoulders + Back  Meals
Coffee
Egg whites sandwich
Donut Chocolate + Orange juice (post workout)
Chicken + Pasta
Protein shake


22-September-2024

15 km run zone 2

Meals: Yogurt
Granola bar
Chicken sandwich (post workout) + electrolytes
Pasta + Shrimp
Protein shake
Chicken sandwich 
Chips  


23-September-2024

Chest
Swim 1300 

Meals: Yogurt
Granola bar
Rice + chicken
Protein shake
2 boiled eggs
Oatmeal w banana and honey 
Rice + chicken + chips 
Milk candy  + bread 


24-September-2024

Back
Sparring sesh


Meals:  Granola Bar
Oatmeal + Banana
1 whites 2 boiled eggs
Sweet bread 
Protein shake 
Apple
Coke no sugar
Chicken + rice
One sincronizada
Chicke sandwich + cookie
2 Tacos
1 bagel


25-September-2024

Rest
5km run fast paced


Meals: 
Oatmeal + Banana
2 whites 2 full boiled eggs
Greek yogurt snack
Decaf coffee
Chicken + Rice
Protein shake
Beef



26-September-2024

Legs
Boxing

Meals:

Jelly sandwich 
Water
Oatmeal
2 boiled eggs
Turkey slices + guacamole
Decaf coffee
Nutella one toasted bread
Kinder delice
Salmon + pasta 
One hot dog
Protein shake
Greek yogurt
Greek yogurt


27-sept-2024

Shoulders + Abs 
Swim

Meals:
Granola bar
Banana
Oatmeal w Half banana
Protein shake
3 boiled eggs + 2 quesadillas
Pasta + beef
Chicken 
Granola bar 
Coke no sugar


28-sept-2024

Boxing
Back


Meals:

Sandwich
Protein shake
Candy beans
Electrolytes
Steak + fries
Protein shake
Chicken with tortilla
Tacos



29-sept-2024

10 km run zone 2

Meals:
 “Complex” breakfast
16 hour fast.

30-sept-2024

Rest 
Chest

Meals:

Greek yogurt
Oatmeal + banana 
3 boiled eggs + quesadilla 
Decaf coffee 
Cake slice (small) 
Bites 
Protein shake 
Beef with tortillas
Greek yogurt + nuts
Beef w tortillas

1-oct-2024

WINTER ARC STARTS

Weight: 73.6

Back
Hard Sparring (5 rounder)

Meals:
Cajeta treat (before workout)
Aminos
Protein shake (24grs)
Decaf
Egg whites sandwich (30 grs)
Soup
Chicken+ vegetables (40 grs)
crepes 
Protein shake (24grs)

Chicken breast + 2 tortillas (30-40grs) 

2-Oct-2024

Intervals
Swimming

Meals: 
Vitamin gummies
Protein shake (24 grs)
2 boiled eggs (22grs)
Oatmeal + paw paw
Greek yogurt + nuts (15grs)
Decaf coffee
Chicken + rice (40grs)
Greek yogurt + nuts(15grs)
Crisps
Chicken sandwich(30 grs)
Protein shake(24grs)
Sincronizada

3-Oct-2024

Legs
Boxing no gloves (HIIT) - burned a total of 1,500 cal

Meals:

Vitamin gummies
Granola bar
Decaf cappuccino coffee
Greek yogurt 
2 boiled eggs + 5 ham slices + guacamole
Chilaquiles + beef
Chicken (480cals) + lasana (120 cals)
Pizza and boneless (700)

4-Oct-2024

Shoulders
Swimming (2000m)

Meals:
Granola bar
Protein shake(24)
Oatmeal + banana
3 boiled eggs(25)
2 quesadillas(10)
Greek yogurt + some paw paw (15)
Chicken + rice (30)
2 chicken tacos homemade

5-Oct-2024

Boxing

Meals
Granola bar
Protein shake
Turkey sandich
Panditas
Brownie 
Yakimeshi chicken
Salmon + tuna roll 
(20)
Greek yogurt(15)
Protein Shake(24)
Beef soup (25-30 grs)
2 chicken tacos homemade(20)


6-oct-2023

10km race (PB)

Meals: i dont know anlot pf junk shit

7-oct-2023
Rest


Meals

Protein shake
Granola bar
Oatmeal 
Vitamin gummies
3 boiled eggs
Chicken + rajas and 2 salmas
Pizza
Crisps


8-oct-2023

More aggressive 
Calorie deficit approach STARTS now.
Sparring



Rest in the morning

Meals:

Protein shake + banana
3 whole Eggs+ 1 white (scrambled)
Chicken w rajas + rice
Protein shake
Coke no sugar



9-oct-2023

Intervals
Swimming (1,500)

Meals:

Granola bar
Protein shake + banana
Vitamin gummies
3 whole Eggs+ 1 white (scrambled)
2 chicken breasts breaded (with some cheese in them)
Greek yogurt
Granola bar
Protein shake


10-oct-2023

Legs 
Boxing

Meals
Granola bar
Protein shake + guayaba
Vitamin gummies
4 boiled eggs
2 quesadillas
Greek yogurt
Greek yogurt
Chicken portion , rib + pasta small portion
Protein shake


11-Oct-2024

Chest/Shoulders
Swim

Meals

Granola bar
Banana
Protein shake
4 eggs + 2 slices of bread
Chicken + beef soup
Chocolate treats
Greek yogurt


12-Oct-2024

Boxing

Meals:
Shitty

13-Oct-2024
6 km run zone 2

Meals:

Hotcakes w Nutella and banana
Egg whites omelette 
Shitty meals


14-Oct-2024

Chest and run

Meals:
Granola bar
Protein shake
3 boiled eggs 1 egg white
Pork meat + potatoes (small portion) 
Greek yogurt + berries

Workout 
Protein Shake


15-Oct-2024

Back
Short bike intervals

Meals

Granola bar
Yogurt
Mini oatmeal Pancakes 
3 egg whites 
4 Turkey ham slices
Protein shake
Cucumber
Arrachera + Salad
Coke no sugar
3 kinder bueno mini treats (120 cals)
Burger 70-30 sirloin, egg 


16-oct-2024

Boxing

Meals
Protein shake
Hotdog
Eggs
Cookie
Protein shake

17-Oct-2024

Legs
Boxing

Meals

2 granola bars
Greek yogurt
Workout
Protein shake
Pasta + chicken
1 sincronizada + some sauce
Cookies treat 
Protein shake
Bread


18-Oct-2024 

Shoulders
Swim

Meals
Granola bar, some orange juice
Nutella w bread 
2 egg whites + turkey ham and some huacamola
2 decaf cappuccinos
1 sugar no coke
Burger (200 cals) + french fries + grill salmon
2 greek yogurts + banaba
3 tacos


19-Oct-2024

Boxing

Meals
Granola bar
Protein shake
Sweet bread
Yogurt
Roasted beef 3 steaks
Tortilla and sauce
2 arrachera tacos
Ramen
Protein shake

20-Oct-2024
Exercise 

10 km zone 2
Back

Meals
Greek yogurt
Protein shake
2 kinder delice
Ham croissant
2 carnitas tacos
1 gordita of carnitas
Protein shake
Greek yogurt


Routine
Workout
Church 
Grandma’s 
Rest 
Study Session (Deep learning)
Business data science


21-Oct-2024

Chest
Swim

Meals

Cookies
Protein shake
3 boiled eggs + 1 egg white 
Greek yogurt
Bowl of soup
Chicken fajitas (medium-big sized portion) + 3 tortillas
Cookies
Greek yogurt
Protein shake (zero carbs)


Routine
Gym (1 hr and a half)
Breakfast
Prep for work
Work 
Short study session (15 min) - bootstrap 
Work
Swimming session
Writing session for my post in medium 1 hr


22-Oct-2024
Boxing
Back

Meals
Granola bar
Decaf coffee cappuccino
Protein shake
Strudel portion
3 boiled eggs
Chicken salad + salmas
Chicken soup ( w a portion of chicken) plus two quesadillas of panela cheese
Cookies
Protein shake (140 cals)


23-Oct-2024


Rest

Meals

Protein shake
Cookies
2 Boiled eggs + 2 egg whites + toasted bread w cheese slice and avocado
vitamin gummies
Coffee little caffeine
Chicken salad + noodles soup
Cookies 
Chicken sandwich + crisps w sauce
Sincronizada
Pamela Cheese quesadilla


24-Oct-2024

Legs
Boxing


Meals

3 egg whites
Turkey ham portion (6 slices)
One nutella bread
Protein shake (zero carbs)
No sugar coke
Arrachera(medium sized portion)+ salad + cooked vegetables 
Fist sized chicken portion
Red velvet cake (150cals)
No sugar coke
Granola bar 
Banana
2 Light chicken tacos
Protein shake


25-Oct-2024

Shoulders
Swim

Meals
Granola bar
Protein shake
3 scrambles eggs
1 slice of bread
Donut
Greek yogurt
Instant pasta ( 440 cals)
Chicken breast (150 grs aprox) w some sauce
Protein shake

26-Oct-2024

Boxing


Meals

Granola bar
Banana
Boneless (1,100 cals)
Tacos 3 
Aztek soup

27-Oct-2024
Rest


Meals
Banana
Protein shake
2 quesadillas
Kinder delice
Subway foot long
Doritos
Sandwich
Donut

28-Oct-2024

Chest
10 km zone 2 run

Meals

Mamut
Protein shake
Decaf coffee no sugar
2 Eggs  2 whites + some bread
Greek yogurt
Chicken nuggets 
Medium french fries
Coke zero 

(1500) 

Protein shake (180)
Chicken slice + mushrooms (300)

29-Oct-2024
Back
Boxing

Meals

Tiny mamut
2 egg whites
4 turkey ham slices
Kinder delice
2 slices of chicken
1 tiny pizza
French fries
Chicken bahn 250 cals 
Sandwich (tuna pieces)
Hojuela de papa (100 cals)
Beef + rice
Protein shake 




30-Oct-2024

Intervals 400x8, 2km
Swim

Meals

Sweet bread x2
4 egg whites
Protein shake
1 slice of bread  + cheese
Coffee
Greek yogurt
2 slices of pork meat 
letters Soup 
Crisps 
Small portion greek yogurt
Protein shake
Sandwich 


31-oct-2024

Boxing

Meals

Protein shake 
2 egg whites 1 egg (boiled)
Sincronizada (4 slices of light turkey ham  + cheese)
1 slice of bread 
Chicken breast + rice + vegetables
Greek yogurt + grapes
Sincronizada
Protein shake
Sweet bread
Soda

1-nov-2024

Rest
Shoulders

Meals
Protein shake
Medium sized Pepperoni pizza pizza hut
Tacos orders

2-Nov-2024

Boxing
Back

Meals

Banana
Protein shake + banana
4 eggs + 2 turkey ham slices
Greek yogurt
Ramen


3-Nov-2024

12km zone 2 run

Meals

Protein shake + banana + greek yogurt spoon
Brownie
250 grs beef portion
Empanada de chistorra
Jugo de carne
Rice + fish fillet (crisp) 120grs
Greek yogurt cup
Cereal

4-Nov-2024

Chest
Swim intervals

Meals
Protein shake
coffee
1 quesadilla
3 eggs
Rice + pork beef
McCombo
Protein shake
Greek yogurt 

From this point on, we start a stronger deficit.


4-Nov-2024

Back
Boxing

Meals

Half banana
Post workout
Protein shake (milk , scoop and banana)
2 egg whites 1 whole egg
1 slice of bread
1 slice of bread + milk candy
Greek yogurt (of strawberry)
Meat (it was a hand a half portion) + some cheese + cauliflower + 2 harina tortilla 
Chicken + rice 
Meat + cheese

5-Nov-2024

Running intervals


Meals

Protein shake (banana + scoop)
2 whole eggs + 1 white + 2 tortillas
Greek yogurt
Coffee
Meat lasaña (one and a half+ calabasas + 2 tortilla
Greek yogurt plus chocolate
Ramen (450 cals)
Sandwich
Protein shake + Milk
2 bread slices + cajeta


6-nov-2024

Swimming 1,500
Boxing


Meals

Protein shake + banana
Greek yogurt
Meat guisado (250rs) + Rice 
Greek yogurt	(arandanos + nuts)
Sincronizada (cheese + turkey ham)
Meat guisado 2 tortillas
Prote


8 nov 2024

Shoulders
Back

Meals
Banana
2 egg whites
5 turkey ham slices
Oatmeal
Kinder delice
Coffe cappucino
250 grs arrachera
Pasta 
French fries
Kinder delice 
Pizza

9-nov-2024

Chest

Meals


Greek yogurt
Chocolate
Bread + sweet milk
Pizza
Bad day of eating

10-nov-2024

13.5 km race
5’30’ pace


Meals

Bad day of eating

11-Nov-2024

Chest


Meals

Coffee 
Greek yogurt (berries and almonds)
Broccoli + chicken + some sauce
Broccoli + chicken + some sauce (smaller portion)
Greek yogurt
Greek yogurt
3 egg whites +  coliflor + cheese
Bread slice
Cake slice

11-Nov-2024

Back
Boxing

Meals

2 chocolates no sugar
Greek yogurt (starverry)
Greek yogurt coconut
Chicken + rice + 2 quesadillas
Greek yogurt
Chicken + vegetables + sauce
Donut + milk 


12-Nov-2024

Intervals
Swimming intervals

Meals

Milk (high in protein )+ banana milkshake
4 egg whites + 2 homemade quesadillas (flour tortilla) 
greek yogurt + banana + nuts
Rice + chicken + some broccoli
Rice  + a chicken leg
Greek yogurt
2 chocolates no sugar
1 quesadilla + ham slice 


13-nov-2024

Legs
Boxing


Meals

Milk (high in protein )+ banana milkshake
3 eggs + 3 ham turkey slices 
Milk + sweet bread
Oatmeal + grapes
Rice + 150 grs beef
2 slices of bread plus milk candy
Rice + chicken 
3 chocolate mini no sugar


14-Nov-2024


HIIT-boxing (1k cals)

Meals

4 eggs + 1 slice of bread
Coffee no sugar + some milk 
1 cup of oatmeal + half banana
Protein shake (whey iso) 
Greek yogurt + half banana
1 chocalate no sugar
2 Eggs whites + rice
1 greek yogurt
Protein shake
1 sincronizada

15-nov-2024
Back
Swimming

Meals
Bad as fuck probably

16-nov-2024
5km zone 2 run


Meals
Went to the cinema
Fucked up

17-nov-2024

Chest

Meals

Scrambled eggs + sausages
Protoein shake
Some cake (like 300 cals prolly)
Protein shake
Greek yogurt
Eggs with some beef
2 chocalates mini no sugar
Greek yogurt

Rice + beef (small portion)


19-nov-2024

3 WEEK PLAN CUT
Back
Boxing

Meals

Protein shake
4 Egg whites boiled 
Oatmeal + 2 chocolates no sugar
Rice + chicken in sauce
3 integral cookies
Greek yogurt
Protein shake


20-nov-2024


Run 3km fast pace 1 km warm up

Swimming intervals

Meals

Protein shake
Coffee
Quesadilla
2 Egg whites boiled  2 full eggs
Oatmeal + cranberry 
Spaghetti + beef in sauce
Cajeta sandwich + some milk
Greek yogurt
Protein shake
Sincronizada


21-nov-2024


Legs
Boxing

Meals

Protein shake + banana
3 scrambled eggs + 3 slices of turkey ham + one slice of integral bread
Oatmeal + cranberry 
Beef + sauce + chicken (80-100grs) + 3 tortillas + some cheese
Soda (150cals)
Greek yogurt
Slice of pizza
Protein shake


22-nov-2024


Shoulders

Meals

Protein shake + banana
1 sincronizada
1 slice of bread + cajeta 
Greek yogurt
Chicken (fist sized portion) + sauce + 3 tortillas + panela cheese + spaguetti
Nature valley bar
Protein shake
Greek yogurt + raisins


23-nov-2024

Boxing

Meals

Slice of bread + cajeta + some milk
Protein shake
Greek yogurt + raisins
Chicken tacos + panela cheese (3)
chocolate cake slice + milk
Protein shake  + banana
Some pasta
Life savers
Chocolate cake slice
Ramen (450 cals)

24-nov-2024

12 km zone 2 run
Back

Meals


Banana
Some cake 
Protein shake +milk
Sandwich
Vegetables soup + chicken

25-nov-2025

Chest
Meals
Not that bad 

26-nov-2025

Rest

Meals
Burger
Tacos
Shitty af

27-nov-2024
Rest
(Got sick)
Meals

Tea
Greek yogurt
Protein shake + banana
2 eggs 1 egg white
2 quesadillas + chicken in fajitas 
3 integral Salmas
Greek yogurt
Sgrimps
Protein



28-nov-2024

Back
HIIT

Meals

Granola bar
Cappucino
3 egg whites + 6 slices of turkey ham
Oatmeal + some paw paw
Greek yogurt
Protein shake
Chicken + roast beef + vegetables 
Bread prepared like a tapa
Rice + some sauce Fish
Red velvet dessert (130 cals)
1.5 slice of pizza
Protein shake

29-Nov-2024

Shoulders
HIIT swimming

Meals

Protein shake + banana
Coffee
Greek yogurt
Oatmeal w milk and water + some honey 
2 boiled eggs
Granola bar
Lasaña bowl (home made) x2
Boneless
Tinga

30-Nov-2024

Back
HIIT BOXING

Meals
 

Greek yogurt
Oatmeal  w banana
Chicken + broccoli + cheese
Protein shake (water)
Chicken + broccoli + cheese (smaller portion
Protein shake + banana
Chocolate cake (small portion)
Sandwich w ham, panela cheese and cheese


1-dic-2024


Zone 2 run


Meals

Greek y
Oatmeal + half banana
Enchiladas suizas
Pan dulce
Cafe descafeinado
Medio emparedado de pollo
Alitas de pollo (6) + papas



2-dic-2024


Chest
Swim intervals

Meals


Protein shake + half banana
Oatmeal + honey
3 Boiled eggs
Greek yogurt
Granola bar
Chicken in tinga sauce + rice (1.5 portion)
Greek yogurt
Granola bar
Protein shake
Sandwich w panela chees

3-dic-2024


Back and biceps
Boxing

Meals


Protein shake + half banana
Granola bar
3 eggs + 3 pork ham slices + 1 slice of bread + a small portion of mayo
Sweet Banana + some candy 
Chicken w sauce + rice
Greek yogurt + some raisins
protein shake
Greek yogurt
Sincronizada


4-dic-2024

Legs
Swimming intervals

Meals

Protein shake + banana + some milk
Greek yogurt
Tea 
3 eggs + 2 quesadillas
Caldo de verduras + 3 beef portions hand sized
Granola bar
Canelitas (3)
Protein shake
Sweet bread

5-dic-2024

Rest

Meals

Protein shake + banana
3 fried eggs 1 slice of bread
Slice of chocolate cake dried + some
Milk
Rice + Chicken (small portion)
Sandwich w panela cheese and cheese + 3 slices of pork ham
2 small packs of canelitas
Cereal
Caldo
Shitty day


6-dic-2024

Shoulders
Biceps and back

Meals

Granola bar
3 egg whites
Oatmeal
Schorle (90 cals)
Salmon + Salad
Tuna + potato baked shit plus some vegetables 
15 boneless only the boneless (750 cals)

7-dic-2024

Chest
Boxing HIIT

Meals

Protein shake + banana 
Oatmeal + raisins
Protein shake 
Burger (350 cals)
2 slices of pizza home made 


8-dic-2024

10km run zone 2
1200 swim

Meals
Granola bar
Protein shake
2 Tacos
Chicken + some pasta
Popcorn/nachos

9-Dic-2024
 6km run zone 3-4
Chest

Meals
Protein shake
Banana
Eggs w sausage
Oatmeal
Yakimeshi + chicken salad

10-Dec-2024

Rest

Meals
Shitty

11-Dec-2024

Fast paced 5km
Back

Meals

Protein shake + banana
3 eggs + 2 slices of ham
Oatmeal + raisins
Beef in sauce + potatoes
Greek yogurt
Protein shake
Greek yogurt
Beef+ sauce + calabaza 
sweet bread

12-Dec-2024

Legs
Boxing (hiit) at home

Meals

Granola bar
Protein shake
3 boiled eggs
Oatmeal + raisins + some honey
Greek yogurt strawberry
2 chicken steaks + chicharron w sauce (2 portions of chicharron)
1 sweet bread
1 protein shake

16-Dec-2024

Chest

Meals
Protein shake + banana
Oatmeal + raisins
Tea
2 chicken steaks breaded + 2 tortillas + 2 slices of panela cheese.
2 granola bars
1 Costco hot dog
1 granola bar
Protein shake


17-Dec-2024

Back
Bike
Boxing

Meals

Granola bar
Some cake (just like 1 spoon)
Costco hotdog
Protein shake
Greek yogurt 
Pechuga empalizada + espagueti + some vegetables in sauce + some pork (really small portion)
2 tortillas w cajeta
Protein shake
Sweet bread


18-Dec-2024

Swim (1,500)
Legs workout

Meals
Granola bar
Some candy
Oatmeal + 2 egg whites + 6 turkey ham slice + 2 slices of panela cheese
1 slice of bread + Nutella 
Decaf cappuccino
Pasta + salmon + pork rib
Cake treat
Kinder delice
Greek yogurt
Nuts bar
Guisado de chuleta + spaguetti 


19-Dec-2024

Shoulders
Bike

Meals 
Protein shake + banana
4 Eggs + 1 slice bread
3 hot cakes
1 milk cup
Guisado de chuleta + spaguetti
Protein shake
Greek yogurt
Sandwich

20-Dec-2024

Back
Run zone 3

Meals
Protein shake + banana
3 fried eggs + 1 slice of integral bread
Orange juice
Granola bar


21-Dec

2000 m swim

Cajeta + 1 slice of bread
Rice + pork guisado

23-Dec-2024
2000m swim
10km run indoor

Meals
Protein shake
Shitty

24-Dec-2024
Chest
Legs

Meals
Protein shake
  """
    category_counts = parse_workout_categories(text_data)

    print("Workout Category -> Frequency")
    for cat, freq in sorted(category_counts.items(), key=lambda x: x[1], reverse=True):
        print(f"{cat}: {freq}")

    plot_category_counts(category_counts)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# 1) Suppose you have your dictionary, but we won't re-paste it all.
#    We'll call it 'cal_dict' for brevity. We'll filter out zeroes.



def parse_date(d_str):
    """
    Attempt to parse strings like '17-August-2024'.
    We'll do a simple fix if '20224' is present -> '2024'
    """
    d_str = d_str.replace("20224", "2024")  # fix obvious typos
    fmts = ["%d-%B-%Y", "%d-%b-%Y"]
    for f in fmts:
        try:
            return datetime.strptime(d_str, f)
        except:
            pass
    return None

# 2) Build a list of (date_obj, cals), ignoring zeros and unparseable dates
date_cal_pairs = []
for d_str, cals in cal_dict.items():
    if cals > 0:  # Skip zero-calorie days
        dt = parse_date(d_str)
        if dt:
            date_cal_pairs.append((dt, cals))

# 3) Sort by date
date_cal_pairs.sort(key=lambda x: x[0])

if not date_cal_pairs:
    print("No data after filtering out zeros!")
else:
    # 4) Convert to arrays for polynomial fitting
    xdates = [p[0] for p in date_cal_pairs]
    ycals = [p[1] for p in date_cal_pairs]

    # Convert xdates to numeric (days from earliest date)
    x0 = xdates[0]
    x_vals = np.array([(d - x0).days for d in xdates], dtype=float)
    y_vals = np.array(ycals, dtype=float)

    # 5) Fit a higher-degree polynomial, e.g. deg=3
    degree = 3
    coefs = np.polyfit(x_vals, y_vals, deg=degree)
    poly_func = np.poly1d(coefs)

    # 6) Generate fitted curve
    y_fit = poly_func(x_vals)

    # 7) Approximate 95% band (±2σ of residuals)
    residuals = y_vals - y_fit
    res_std = np.std(residuals)
    upper = y_fit + 2 * res_std
    lower = y_fit - 2 * res_std

    # 8) Compute average of non-zero cals
    avg_cals = y_vals.mean()

    # 9) Plot
    sns.set(style="whitegrid", font_scale=1.1)
    plt.figure(figsize=(10, 6))

    # Plot actual data points
    plt.plot(xdates, y_vals, 'o-', label="Daily Calories", color='blue', alpha=0.7)
    
    # Polynomial fit line
    plt.plot(xdates, y_fit, 'r--', linewidth=2, label=f"Poly (deg={degree})")

    # Fill ±2σ band
    plt.fill_between(xdates, lower, upper, color='red', alpha=0.1, label="~95% band")

    # Horizontal average line
    plt.axhline(avg_cals, color='green', linestyle=':', linewidth=2,
                label=f"Average = {avg_cals:.1f} cals")

    # Labels, legend, formatting
    plt.title("Daily Calorie Intake (Non-zero) with Polynomial Trend & 95% Band", fontsize=14)
    plt.xlabel("Date")
    plt.ylabel("Calories")
    plt.legend()
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```
